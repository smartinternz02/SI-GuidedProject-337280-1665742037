```python
from tensorflow.keras.layers import Dense, Flatten, Input
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img
from tensorflow.keras.applications.vgg19 import VGG19, preprocess_input
from tensorflow.keras.callbacks import EarlyStopping
from glob import glob
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

```python
imagesize = [224,224]

data =r"/content/drive/MyDrive/Yoga Pose/Dataset"
```

```python
vgg = VGG19(include_top=False, input_shape=imagesize+[3],weights="imagenet")
```

```python
for layer in vgg.layers:
  layer.trainable = False
```

```python
x = Flatten()(vgg.output)
```

```python
prediction = Dense(5, activation='softmax')(x)
```

```python
model = Model(inputs = vgg.input, outputs = prediction)
```

```python
model.summary()
```

```python
model.compile(
    loss = 'categorical_crossentropy',
    optimizer = 'adam',
    metrics = ['accuracy'], run_eagerly = True
    )
```

```python
train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True,
                                   validation_split = 0.3)
```

```python
training_set = train_datagen.flow_from_directory(data,
                                                 target_size = (224,224),
                                                 batch_size = 8,
                                                 class_mode = 'categorical',
                                                 subset='training')

testing_set = train_datagen.flow_from_directory(data,
                                                target_size = (224,224),
                                                batch_size = 8,
                                                class_mode = 'categorical',
                                                subset='validation')

```

```python
training_set.class_indices
```

```python
len(training_set)
```

```python
r = model.fit(
    training_set,
    validation_data = testing_set,
    epochs = 20,
    steps_per_epoch = len(training_set)//8,
    validation_steps = len(testing_set)//8,
)
```

```python

model.save('vgg19_yoga.h5')
```
